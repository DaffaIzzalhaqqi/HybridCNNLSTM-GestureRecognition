```python
import cv2
import os
import time

# Define the ROI dimensions
roi_size = 500
x = int((1000 - roi_size) / 2)
y = int((1000 - roi_size) / 2)
w = roi_size
h = roi_size

# Define the number of images to capture per class
num_images = 1000

# Define the classes to capture images of
classes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I',
          'K', 'L', 'M', 'N', 'O', 'P', 'Q', 'S', 'T',
          'U', 'V', 'W', 'X', 'Z']

# Define directory
save_dir = '/Users/daffaizzalhaqqi/Desktop/Seminar Proposal Skripsi Daffa/Skripsi/Data Collection/Training'

# Set up the camera
cap = cv2.VideoCapture(0)

# Create the directories for the images
for c in classes:
    os.makedirs(os.path.join(save_dir, c), exist_ok=True)

# Loop over the classes
for c in classes:
    
    # Loop over the number of images to capture
    for i in range(num_images):
        
        # Capture the frame
        ret, frame = cap.read()

        # Crop the frame to the ROI
        roi = frame[y:y+h, x:x+w]

        # Make a copy of the ROI
        roi_with_text = roi.copy()

        # Define the message to display
        message = "Capturing images of {}".format(c)
        current_state = "{}/{}".format(i+1, num_images)

        # Display the message and current state on the copy of the ROI
        cv2.putText(roi_with_text, message, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        cv2.putText(roi_with_text, current_state, (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        
        if i == 0:
            # Pause for 10 seconds before capturing the first image of each class
            time.sleep(10)

        # Display the ROI
        cv2.imshow("ROI", roi_with_text)

        # Define the path and filename for the image
        path = os.path.join(save_dir, c, "{}_{}.png".format(c, i))

        # Write the original ROI to disk
        cv2.imwrite(path, roi)

        # Pause for 0.01 seconds before capturing the next image
        time.sleep(0.01)

        # Check for key press to exit the program
        if cv2.waitKey(1) == 27:
            break

    # Close the window
    cv2.destroyAllWindows()

# Release the camera
cap.release()
```
